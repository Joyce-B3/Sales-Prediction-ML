```python
import pandas as pd  
import matplotlib.pyplot as plt
# Charger les fichiers CSV (modifie le nom si nécessaire)
train_df = pd.read_csv("C:\\Users\\joyce\\Desktop\\mesdocuments\\data science\\projet\\Sales-Prediction-ML\\data\\train.csv") 
test_df = pd.read_csv("C:\\Users\\joyce\\Desktop\\mesdocuments\\data science\\projet\\Sales-Prediction-ML\\data\\test.csv")
oil_df = pd.read_csv("C:\\Users\\joyce\\Desktop\\mesdocuments\\data science\\projet\\Sales-Prediction-ML\\data\\oil.csv")
stores_df = pd.read_csv("C:\\Users\\joyce\\Desktop\\mesdocuments\\data science\\projet\\Sales-Prediction-ML\\data\\stores.csv")
transactions_df = pd.read_csv("C:\\Users\\joyce\\Desktop\\mesdocuments\\data science\\projet\\Sales-Prediction-ML\\data\\transactions.csv")
holidays_df = pd.read_csv("C:\\Users\\joyce\\Desktop\\mesdocuments\\data science\\projet\\Sales-Prediction-ML\\data\\holidays_events.csv")

# Aperçu des données
print(train_df.shape)  # Vérifier la taille
print(train_df.head())  # Voir les premières lignes
print(test_df.head())  # Voir les premières lignes
print(train_df.info())  # Vérifier les types de données
```

```python
print(train_df.isnull().sum())  # Vérifier les valeurs manquantes dans train.csv
print(test_df.isnull().sum())  # Vérifier les valeurs manquantes dans test.csv
print(oil_df.isnull().sum())  # Vérifier les valeurs manquantes dans oil.csv
print(transactions_df.isnull().sum())  # Vérifier les valeurs manquantes dans transactions.csv
print(holidays_df.isnull().sum())  # Vérifier les valeurs manquantes dans holidays_events.csv
```

```python
oil_df['dcoilwtico'] = oil_df['dcoilwtico'].fillna(oil_df['dcoilwtico'].median())
```

```python
print(oil_df.isnull().sum())  # Vérifier si tout est OK
```

```python
train_df = train_df.drop_duplicates()
```

```python
print(train_df.duplicated().sum())  # Nombre de doublons
```

```python
train_df['date'] = pd.to_datetime(train_df['date'])
test_df['date'] = pd.to_datetime(test_df['date'])
oil_df['date'] = pd.to_datetime(oil_df['date'])
transactions_df['date'] = pd.to_datetime(transactions_df['date'])
holidays_df['date'] = pd.to_datetime(holidays_df['date'])
```

```python
train_df['year'] = train_df['date'].dt.year
train_df['month'] = train_df['date'].dt.month
train_df['day'] = train_df['date'].dt.day
train_df['day_of_week'] = train_df['date'].dt.dayofweek  # Lundi=0, Dimanche=6
```

```python
plt.figure(figsize=(12, 6))
train_df.groupby('date')['sales'].sum().plot()  
plt.title('Ventes totales par jour')
plt.xlabel('Date')
plt.ylabel('Ventes')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
train_df.groupby('store_nbr')['sales'].sum().plot(kind='bar') 
plt.title('Ventes totales par magasin')
plt.xlabel('Magasin')
plt.ylabel('Ventes')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(train_df["sales"], bins=50, kde=True)
plt.xlabel("Ventes")
plt.ylabel("Fréquence")
plt.title("Distribution des ventes")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=train_df["sales"])
plt.title("Boxplot des ventes")
plt.show()
```

```python
import seaborn as sns
selected_columns = ['sales', 'id', 'store_nbr', 'date','onpromotion',]
# Calculer la matrice de corrélation
corr_matrix = train_df[selected_columns].corr()

# Afficher la heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matrice de corrélation")
plt.show()
```

```python
# Afficher la heatmap avec un meilleur contraste
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)
plt.title("Matrice de corrélation ajustée")
plt.show()
```

```python
# Exemple de scatterplot entre 'sales' et 'transactions'
plt.figure(figsize=(8,6))
sns.scatterplot(x=transactions_df['transactions'], y=train_df['sales'])
plt.title("Relation entre les transactions et les ventes")
plt.show()
```
